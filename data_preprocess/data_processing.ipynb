```python
import math
def eucliDist(A, B):
    """Calculate the A,B Euclidean distance"""
    return math.sqrt(sum(np.power((A - B), 2)))
```

```python
def getLength(item):
    """Calculate road distance"""
    euc_arr = []
    for i in range(0,len(item)-1):
        (x_from, y_from), (x_end, y_end) = item[i], item[i+1]
        A, B = np.array((x_from, y_from)), np.array((x_end, y_end))
        length = eucliDist(A, B)
        euc_arr.append(length)
    # print(sum(euc_arr))
    return sum(euc_arr)
```

```python
from shapely.geometry import LineString
temp = []
cost = []
for idx, row in small_wes.iterrows():
    item = list(row['geometry'].coords)
    # Selecting the center as the route coordinates
    x_start, y_start = item[0]
    x_end, y_end = item[len(item)-1]
    x, y = (x_start + x_end)/2, (y_start + y_end)/2
    coordinate = (x, y)
    temp.append(coordinate)
    # Calculate route distance
    euclen = getLength(item)
    cost.append(euclen)


small_wes['coordinate'] = temp
small_wes['newcost'] = cost
small_wes
```

```python
small_wes = small_wes[['from', 'to', 'newcost', 'coordinate']]
small_wes
```

```python
small_wes = small_wes.reset_index(drop=True)
small_wes
```

```python
coord_list = list(small_wes['coordinate'])
# eucdist_list = list(small_wes['newcost'])

from_list = []
to_list = []
eucdist_list = []
euc_distance_list = []
for idx_i, i in enumerate(coord_list):
    for idx_j, j in enumerate(coord_list):
        if idx_i < idx_j:
            euc_distance = eucliDist(np.array(i),np.array(j))
            euc_distance_list.append(euc_distance)
            if euc_distance <= 2000:
                # print(f'{idx_i}->{idx_j}:',euc_distance)
                from_list.append(idx_i)
                to_list.append(idx_j)
                eucdist_list.append(euc_distance) 
euc_distance_mean = np.mean(euc_distance_list)
euc_distance_median = np.median(euc_distance_list)
print("The median of 'euc_distance' is:", euc_distance_median)
print("The average of 'euc_distance' is:", euc_distance_mean)
# percentile
q1 = np.percentile(euc_distance_list, 25)
q3 = np.percentile(euc_distance_list, 75)
print("The first quartile of 'euc_distance' is:", q1)
print("The third quartile of 'euc_distance' is:", q3)
# min and max 
euc_distance_max = np.max(euc_distance_list)
euc_distance_min = np.min(euc_distance_list)
print("The maximum of 'euc_distance' is:", euc_distance_max)
print("The minimum of 'euc_distance' is:", euc_distance_min)

new_wes = pd.DataFrame()
new_wes['from'] = from_list
new_wes['to'] = to_list
new_wes['cost'] = eucdist_list
new_wes
```
